```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# === 1. Load cleaned dataset ===
df = pd.read_csv("cleaned_combined_dataset.csv")
display(df.head())

# === 2. Feature engineering: price difference ===
df['price_diff'] = df['raw_price'] - df['current_price']

# === 3. Normalize selected features ===
score_feats = ['likes_count', 'discount', 'price_diff']
scaler = MinMaxScaler()
norm = pd.DataFrame(
    scaler.fit_transform(df[score_feats]),
    columns=[f"{c}_norm" for c in score_feats],
    index=df.index
)
df = pd.concat([df, norm], axis=1)

# === 4. Compute Product Score ===
df['product_score'] = (
    0.6 * df['likes_count_norm'] +
    0.3 * df['discount_norm'] +
    0.1 * df['price_diff_norm']
)

# === 5. Identify Top 10 products (overall) ===
top_10_products = df.sort_values('product_score', ascending=False).head(10)
print("=== Top 10 Products (overall) ===")
print(top_10_products[['category', 'current_price', 'likes_count', 'discount', 'product_score']])

# === 6. Compute Best Category ===
category_scores = df.groupby('category')['product_score'].mean().sort_values(ascending=False)
print("\n=== Best Category by Average Product Score ===")
print(category_scores)

# === 7. Save results (optional, for Task 4 report) ===
top_10_products.to_csv("top_10_products.csv", index=False)
category_scores.to_csv("best_category_scores.csv")
```
